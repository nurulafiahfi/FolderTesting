```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
# import record set dari file csv
data = pd.read_csv('D:\Dataset\Market_Basket_Optimisation.csv', header = None)
```

```python
import numpy as np
# membuat basket transaction
transaction = []

# memasukkan record set ke dalam basket transaction
for subscript1 in range(0, 7501):
    transaction.append([str(data.values[subscript1, subscript2]) for subscript2 in range(0, 20)])
```

```python
# mengkonversi basket transaction menjadi bentuk array
transaction = np.array (transaction)
transaction
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()

# mengkonversi basket transaction dalam bentuk array menjadi bentuk boolean value
te_bool = te.fit(transaction).transform(transaction)
te_bool
```

```python
# diubah menjadi data frame
data_frame = pd.DataFrame(te_bool, columns = te.columns_)
data_frame
```

```python
data_frame.shape
```

```python
# menghapus kolom nan dalam data frame
data_frame = data_frame.drop("nan", 1)
data_frame.shape
```

```python
# menemukan frequent itemset yang memenuhi minimal support

frequent_itemsets = apriori(data_frame, min_support = 0.04, use_colnames=True)
frequent_itemsets['length'] = frequent_itemsets['itemsets'].apply(lambda x: len(x))
frequent_itemsets
```

```python
# Memilih frequent Itemset yang memiliki tingkat kepercayaan (confidence) 
# dan lift (hubungan antara antecedent dan consequent ) yang tinggi
from mlxtend.frequent_patterns import association_rules
rules = association_rules (frequent_itemsets, metric = 'lift', min_threshold = 1)
rules.sort_values('confidence', ascending = False, inplace = True)
rules
```
